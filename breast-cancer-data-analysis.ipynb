```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
dataset.diagnosis = encoder.fit_transform(dataset.diagnosis)
```

```python
dataset.head()
```

```python
data_M = dataset[dataset.diagnosis == 1]
data_B = dataset[dataset.diagnosis != 1]
```

```python
correlation = dataset.corr()
corr_array = np.array(correlation)
corr_cols = correlation.columns.tolist()
```

```python
import plotly.offline as pyo
from plotly.offline import init_notebook_mode
import plotly.graph_objs as go
```

```python
init_notebook_mode(connected=True)
```

```python
# Analyzing the correlation between the attributes

trace_1 = go.Heatmap(
    z = corr_array,
    x = corr_cols,
    y = corr_cols
)

layout = go.Layout(
    {'title': 'Correlation Matrix of the Attributes'},
     autosize = False,
     height  = 720,
     width   = 800,
     margin  = dict(r = 0 ,l = 210,
                    t = 25,b = 210,),
     yaxis   = dict(tickfont = dict(size = 9)),
     xaxis   = dict(tickfont = dict(size = 9)),
)

fig = go.Figure(data=[trace_1], layout=layout)
pyo.iplot(fig)
```

```python
def plot_corr(feat_1, feat_2):
    
    """ function for plotting the positive and negative correlated features """
    
    trace_0 = go.Scatter(
        x = data_M[feat_1],
        y = data_M[feat_2],
        name = 'Malignant',
        mode = 'markers'
    )
    
    trace_1 = go.Scatter(
        x = data_B[feat_1],
        y = data_B[feat_2],
        name = 'Benign',
        mode = 'markers'
    )
    
    visdata = [trace_0, trace_1]
    
    layout = go.Layout(
        {'title': feat_1 + ' vs ' + feat_2},
        xaxis = {'title': feat_1},
        yaxis = {'title': feat_2}
    )
    
    fig = go.Figure(data=visdata, layout=layout)
    pyo.iplot(fig)
```

```python
data_x = dataset.drop(['diagnosis'], axis=1)
y = dataset.diagnosis
```

## Feature importance and classification using Random Forest

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_x, y, test_size = 0.2)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
classifier = RandomForestClassifier(n_estimators=500)
classifier.fit(X_train_scaled, y_train)
```

```python
# getting the feature importance from random forest classifier

importance = classifier.feature_importances_
indices = np.argsort(importance)[::-1]
```

```python
feature_labels = list(X_train.columns)
```

```python
# plotting the features and their feature importance

trace_0 = go.Bar(
    x = [feature_labels[i] for i in range(X_train_scaled.shape[1])],
    y = importance[indices]
)

layout = go.Layout(
    {'title': 'Feature Importance'},
    xaxis = {'title': 'features', 'tickfont': {'size': 9}},
    yaxis = {'title': 'feature importance', 'tickfont': {'size': 9}}
)

data = [trace_0]
fig = go.Figure(data=data, layout=layout)
pyo.iplot(fig)
```

```python
# taking top 10 important features

features = [
    'radius_mean', 
    'texture_mean', 
    'perimeter_mean', 
    'area_mean', 
    'smoothness_mean', 
    'compactness_mean', 
    'concavity_mean',
    'concave points_mean',
    'symmetry_mean',
    'fractal_dimension_mean'
]
```

```python
final_data = data_x[features]
final_data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(final_data, y, test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
classifier = RandomForestClassifier(n_estimators=500)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn import metrics
```

```python
# model evaluation

print(metrics.roc_auc_score(y_test, y_pred))
```

```python
print(metrics.classification_report(y_test, y_pred))
```

## Feature extraction using PCA

```python
from xgboost import XGBClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_x, y, test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# using PCA for dimensionality reduction
# taking top 10 features

pca = PCA(n_components=10, svd_solver='full')
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
pca.explained_variance_ratio_
```

```python
classifier = XGBClassifier(n_estimators=100)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(metrics.roc_auc_score(y_test, y_pred))
```

```python
print(metrics.classification_report(y_test, y_pred))
```
